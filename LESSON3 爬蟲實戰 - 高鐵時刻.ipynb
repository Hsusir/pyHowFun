```python
# 台北->台南, 2018/02/03, 05:30後啟程的班車

import requests
from bs4 import BeautifulSoup

url = 'https://www.thsrc.com.tw/tw/TimeTable/SearchResult'

# 找到option值
form_data = {
    'StartStation': '977abb69-413a-4ccf-a109-0272c24fd490',
    'EndStation':'9c5ac6ca-ec89-48f8-aab0-41b738cb1814',
    'SearchDate':'2018/02/03',
    'SearchTime':'05:30',
    'SearchWay':'DepartureInMandarin'
}

response_post = requests.post(url, data=form_data)

soup_post = BeautifulSoup(response_post.text, 'lxml')
soup_post
```

```python
# 所有班車(train_number)
td_col1 = soup_post.find_all('td', {'class': 'column1'})

train_numbers = []

for tag in td_col1:
    train_numbers.append(tag.text)
    
print(train_numbers)
```
